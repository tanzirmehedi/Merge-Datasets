```python
import pandas as pd

column_name = 'Package'

df1 = pd.read_csv('https://raw.githubusercontent.com/tanzirmehedi/Merge-Datasets/main/2023-10-15.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/tanzirmehedi/Merge-Datasets/main/Agg_RS_full_2.csv')
df3 = pd.read_csv('https://raw.githubusercontent.com/tanzirmehedi/Merge-Datasets/main/Snyk_Packages.csv')

common_values = set(df1[column_name]) & set(df2[column_name]) & set(df3[column_name])

filtered_df1 = df1[df1[column_name].isin(common_values)]
filtered_df2 = df2[df2[column_name].isin(common_values)]
filtered_df3 = df3[df3[column_name].isin(common_values)]

result_df = pd.concat([filtered_df1, filtered_df2, filtered_df3], axis=1)

output_csv_path = '/content/result_end.csv'

result_df.to_csv(output_csv_path, index=False)

print(f"Rows with common values in '{column_name}' from the three CSV files saved to 'result.csv'.")
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/tanzirmehedi/Merge-Datasets/main/2023-10-15.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/tanzirmehedi/Merge-Datasets/main/Agg_RS_full_2.csv')
df3 = pd.read_csv('https://raw.githubusercontent.com/tanzirmehedi/Merge-Datasets/main/Snyk_Packages.csv')

merged_df = pd.merge(df1, df2, on='Package', how='inner')
merged_df = pd.merge(merged_df, df3, on='Package', how='inner')

merged_df = merged_df[['Package', 'Overall', 'FinalRiskScore', 'Score', 'Id', 'Label', 'Degree', 'Pageranks', 'Eigencentrality', 'In_reach']].drop_duplicates('Package')

merged_df.to_csv('merged_file.csv', index=False)

print("Merged CSV file has been created")
```
